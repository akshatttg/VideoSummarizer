```python
import os
import shutil  # Added to handle file operations
import imagehash
from PIL import Image
import numpy as np

## Creating the New dataset for images and transcript 

def alpharemover(image):
    if image.mode != 'RGBA':
        return image
    canvas = Image.new('RGBA', image.size, (255,255,255,255))
    canvas.paste(image, mask=image)
    return canvas.convert('RGB')

def with_ztransform_preprocess(hashfunc, hash_size=8):
    def function(path):
        image = alpharemover(Image.open(path))
        image = image.convert("L").resize((hash_size, hash_size), Image.ANTIALIAS)
        data = image.getdata()
        quantiles = np.arange(100)
        quantiles_values = np.percentile(data, quantiles)
        zdata = (np.interp(data, quantiles_values, quantiles) / 100 * 255).astype(np.uint8)
        image.putdata(zdata)
        return hashfunc(image)
    return function

dhash_z_transformed = with_ztransform_preprocess(imagehash.dhash, hash_size=8)

# Path to the images folder
images_folder = '/kaggle/working'
unique_images_folder = '/kaggle_unique_images'  # New folder for unique images

# Create the unique images folder if it doesn't exist
os.makedirs(unique_images_folder, exist_ok=True)

# Create a set to store unique hashes
unique_hashes = set()
count = 0

# Iterate through image files in the images folder
f1 = open("uniq.txt", "a")
f2 = open("dup.txt", "a")
list1=os.listdir(images_folder) 
list1.sort()
for filename in list1:
    if filename.lower().endswith(('.png', '.jpg', '.jpeg', '.gif', '.bmp')):
        image_path = os.path.join(images_folder, filename)
        hash_value = dhash_z_transformed(image_path)
        
        # Check if the hash is already in the set (redundant image)
        if hash_value in unique_hashes:
            f2.write(f"Redundant Image: {filename}, Hash: {hash_value}\n")
        else:
            temp=image_path
            temp=temp.replace('/kaggle/working/', '')
            temp=temp.replace('.png','')
            YourTranscript=MyTranscript(image_path)
            if(YourTranscript):
                new_row={'Frame':(image_path), 'Transcript': YourTranscript , 'Num': int(temp)}
                df.loc[len(df)]=new_row
#             new_row={'Frame':(image_path), 'Transcript': YourTranscript }
#             df.loc[len(df)]=new_row 
            unique_hashes.add(hash_value)
            f1.write(f"Unique Image: {filename}, Hash: {hash_value}\n")
            count += 1

            # Move the unique image to the unique images folder
            unique_image_path = os.path.join(unique_images_folder, filename)
#             shutil.move(image_path, unique_image_path)

f1.close()
f2.close()
print(f"{count} images found unique and the images unique are in /kaggle/unique_images")
```

```python
# Cluster the embeddings (example with K-means)
from sklearn.cluster import KMeans

num_clusters = 3  # Choose the number of clusters
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
cluster_labels = kmeans.fit_predict(summary_embeddings.numpy())

# Generate summaries for each cluster
summaries = {}
for cluster_id in range(num_clusters):
    cluster_indices = (cluster_labels == cluster_id)
    cluster_text = data.loc[cluster_indices, 'Combined_Text'].tolist()
    cluster_summary = " ".join(cluster_text[:5])  # Choose the first 5 sentences as the summary
    summaries[cluster_id] = cluster_summary

# Calculate average summary length
average_summary_length = sum(len(summary.split()) for summary in summaries.values()) // num_clusters

# Limit summaries for each cluster to the average length
limited_summaries = {}
for cluster_id, summary in summaries.items():
    summary_tokens = tokenizer.encode(summary, max_length=average_summary_length, truncation=True)
    limited_summaries[cluster_id] = tokenizer.decode(summary_tokens)

# Print limited summaries for each cluster
for cluster_id, summary in limited_summaries.items():
    print(f"Cluster {cluster_id} Summary: {summary}")
```
